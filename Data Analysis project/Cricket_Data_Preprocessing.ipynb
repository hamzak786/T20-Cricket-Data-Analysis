```python
#importing necessary libraries
import pandas as pd 
import json 
```

(1) Process the Match result

```python
with open(r"C:\Users\ASUS\Downloads\Data Analysis project\t20_json_files\t20_json_files\t20_wc_match_results.json") as f:
    data = json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis=1, inplace= True)
df_match.head()
```

```python
match_id_dict={}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_id_dict[key1] = row["match_id"]
    match_id_dict[key2] = row["match_id"]

match_id_dict  
```

# **Batting Summary**

```python
with open(r"C:\Users\ASUS\Downloads\Data Analysis project\t20_json_files\t20_json_files\t20_wc_batting_summary.json") as f:
    data = json.load(f)
    all_rec = []
    for rec in data:
        all_rec.extend(rec['battingSummary'])
batting=pd.DataFrame(all_rec)
batting.head(12)
```

```python
batting["out/not_out"]= batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not _out")
batting.head(12)
```

```python
batting.drop(columns = ["dismissal"], inplace = True)
batting.head(10)
```

```python
batting["batsmanName"]= batting['batsmanName'].apply(lambda x: x.replace('â€',''))
batting["batsmanName"]= batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
batting.head(12)
```

Joined the Match result And Batting Summary

```python
batting['match_id']= batting['match'].map(match_id_dict)
batting.head(10)
```

```python
batting.to_csv(r"C:\Users\ASUS\Downloads\Data Analysis project\t20_csv_files\temp.csv", index = False)
```
